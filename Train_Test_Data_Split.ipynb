```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import keras
```

```python
data = []

directory = os.fsencode('/Users/brunoleme/Downloads/Text - Processos Ordinarios')
for i, file in enumerate(os.listdir(directory)):
    filename = os.path.join(os.fsdecode(directory), os.fsdecode(file))
    with open(filename, 'r', encoding='latin1') as f:
        data.append([i, f.read(), 1])

directory = os.fsencode('/Users/brunoleme/Downloads/Text - Processos Sumários')
for j, file in enumerate(os.listdir(directory)):
    filename = os.path.join(os.fsdecode(directory), os.fsdecode(file))
    with open(filename, 'r', encoding='latin1') as f:
        data.append([j + i, f.read(), 0])
        
df_data = pd.DataFrame(data, columns=['Id','Doc','FlgOrdinario'])
```

```python
df_data
```

```python
df_train, df_test = train_test_split(df_data, test_size=0.3, random_state=4)
```

```python
word_index = {}
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2  # unknown
word_index["<UNUSED>"] = 3
for d in [keras.preprocessing.text.text_to_word_sequence(doc) for doc in df_train.Doc.apply(lambda d: str(d).lower())]:
    for w in d:
        if w not in word_index:
            word_index[w] = len(word_index)
```

```python
inv_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
df_train_balanced = pd.concat([df_train[df_train.FlgOrdinario == 1],df_train[df_train.FlgOrdinario == 0].sample(sum(df_train.FlgOrdinario))])
from sklearn.utils import shuffle
df_train_balanced = shuffle(df_train_balanced)
```

```python
X_train = [keras.preprocessing.text.text_to_word_sequence(doc) for doc in df_train.Doc.apply(lambda d: str(d).lower())]
X_train = [[word_index[w] for w in d] for d in X_train]
Y_train = np.array(df_train.FlgOrdinario)
X_train_balanced = [keras.preprocessing.text.text_to_word_sequence(doc) for doc in df_train_balanced.Doc.apply(lambda d: str(d).lower())]
X_train_balanced = [[word_index[w] for w in d] for d in X_train_balanced]
Y_train_balanced = np.array(df_train_balanced.FlgOrdinario)
X_test = [keras.preprocessing.text.text_to_word_sequence(doc) for doc in df_test.Doc.apply(lambda d: str(d).lower())]
X_test = [[word_index[w] if w in word_index else word_index["<PAD>"] for w in d] for d in X_test]
Y_test = np.array(df_test.FlgOrdinario)
```
