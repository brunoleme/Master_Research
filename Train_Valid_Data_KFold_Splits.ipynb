```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score
import pandas as pd
import pickle
```

```python
with open('/Users/brunoleme/Google Drive/Data Master/X_train_final', 'rb') as file:
    X_train_balanced = pickle.load(file)

with open('/Users/brunoleme/Google Drive/Data Master/Y_train_final', 'rb') as file:
    Y_train_balanced = pickle.load(file)
```

```python
skf = StratifiedKFold(n_splits=5, random_state=12345, shuffle=True)
skf.get_n_splits(X_train_balanced, Y_train_balanced)
```

```python
n_split = 1

for train_index, valid_index in skf.split(X_train_balanced, Y_train_balanced):
    
    print(train_index, valid_index)
    
    
    with open('/Users/brunoleme/Google Drive/Data Master/train_index_final_split_' + str(n_split), 'wb') as file:
        pickle.dump(train_index, file)

    with open('/Users/brunoleme/Google Drive/Data Master/valid_index_final_split_' + str(n_split), 'wb') as file:
        pickle.dump(valid_index, file)
        
    n_split += 1
```
